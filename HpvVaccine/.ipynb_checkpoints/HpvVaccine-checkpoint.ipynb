```python
def generate_vax_rate(populations,vaccine):
    vax_populations = []
    cur = 0
    for i in range(len(vaccine)):
        cur+= vaccine[i]
        vax_populations.append(cur)
    vax_rate = np.array(vax_populations[2:]) / np.array(populations)
    return vax_rate

total_populations = np.array(nassau_populations)+np.array(suffolk_populations)
grouped = df.groupby('vax_year').size().reset_index(name='Count')
categories = list(grouped['vax_year'])
values = list(grouped['Count'])
print(values)
vax_rate_total= generate_vax_rate(total_populations,values)
vax_rate_total
```
