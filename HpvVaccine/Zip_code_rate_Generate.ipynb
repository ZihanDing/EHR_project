```python
zips=df_zipcode['zip']
```

```python
def calamount(df,years,start,end):
    rate = []
    for year in years:
        lst = list(df[year])
        count = 0
        for i in range(len(lst)):
            if lst[i]:
                if lst[i]<=end and lst[i]>=start:
                    count += 1
#     print(count)
        rate.append(count)
    return rate
```

```python

def ZipCalAmount(df,years,zips,start,end):
    
    zip_val = {}
    for zp in zips:
        tmp_df = df[df['PatientZip']==zp]
        amount = calamount(tmp_df,years,start,end)
        zip_val[zp] = np.array(amount)
    return zip_val
```

```python
years = np.array((df_age_total.columns))[6:]
```

```python
zp_hpv_amount = ZipCalAmount(df_age_hpv,years,zips,9,13)
```

```python
zp_total_amount = ZipCalAmount(df_age_total,years,zips,9,13)
```
